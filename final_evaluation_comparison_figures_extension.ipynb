```python
import matplotlib.pyplot as plt
import numpy as np

models = ['SVM + HOG', 'ResNet18', 'ViT-B/16', 'SimCLR']
x = np.arange(len(models))
bar_width = 0.35

# Accuracy (mean  std)
train_acc_mean = [0, 99.87, 96.78, 95.72]
train_acc_std = [0, 0.19, 0.12, 0.11]


val_acc_mean = [96.05, 99.53, 98.14, 97.36]
val_acc_std = [0.00, 0.19, 0.00, 0.29]


# Loss (mean  std)
train_loss_mean = [0, 13.25, 19.07, 20.78]
train_loss_std = [0, 0.53, 0.08, 0.34]


val_loss_mean = [0, 12.72, 16.61, 17.84]
val_loss_std = [0, 0.32, 0.05, 0.53]

training_time = [84, 455, 924, 2273]

# the Function to label bars with mean  std
def annotate_bars(ax, bars, stds):
    for bar, std in zip(bars, stds):
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2,
            height + std + 1,
            f'{height:.2f}{std:.2f}',
            ha='center', va='bottom', fontsize=9)

# plot accuracy
fig, ax = plt.subplots(figsize=(10, 5))
bars1 = ax.bar(x - bar_width/2, train_acc_mean, bar_width, yerr=train_acc_std, label='Train Accuracy', capsize=9)
bars2 = ax.bar(x + bar_width/2, val_acc_mean, bar_width, yerr=val_acc_std, label='Validation Accuracy', capsize=9)
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylabel('Accuracy (%)')
ax.set_title('')
ax.set_ylim(0, 105)
ax.legend(loc='lower left')
annotate_bars(ax, bars1, train_acc_std)
annotate_bars(ax, bars2, val_acc_std)
plt.tight_layout()
plt.show()

# plot loss
fig, ax = plt.subplots(figsize=(10, 5))
bars1 = ax.bar(x - bar_width/2, train_loss_mean, bar_width, yerr=train_loss_std, label='Train Loss', capsize=5)
bars2 = ax.bar(x + bar_width/2, val_loss_mean, bar_width, yerr=val_loss_std, label='Validation Loss', capsize=5)
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylabel('Loss (%)')
ax.set_title('')
ax.legend(loc='upper left')
annotate_bars(ax, bars1, train_loss_std)
annotate_bars(ax, bars2, val_loss_std)
ax.set_ylim(0, max(max(train_loss_mean), max(val_loss_mean)) + 3)
plt.tight_layout()
plt.show()

# plot trainig time
fig, ax = plt.subplots(figsize=(10,5))
bars = ax.bar(x, training_time, bar_width, color = 'skyblue')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylabel('Training time (seconds)')
ax.set_title('')
for bar, time in zip(bars, training_time):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height()+20, f'{time}s', ha='center', va='bottom',fontsize=9)
ax.set_ylim(0, max(training_time)+100)
plt.tight_layout()
plt.show()
```
